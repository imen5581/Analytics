Читаем Dataframe

```python
import pandas as pd
df = pd.read_csv('PlayerAttributeData.csv')
df.head(5)
```

Очищаем данные

```python
df['Aggression'] = df['Aggression'].str.extract(r'^(\d{2})', expand=False)
df['Penalties'] = df['Penalties'].str.extract(r'^(\d{2})', expand=False)
df['Aggression'] = pd.to_numeric(df['Aggression'])
df['Penalties'] = pd.to_numeric(df['Penalties'])
df = df[df['Aggression']>70]

```

Строим визуализацию

```python
import plotly.graph_objects as go

f = go.FigureWidget()
f.add_scatter(y=df['Aggression'][:20])
f.add_bar(y=df['Penalties'][:20])
f.layout.title = 'Зависимость пенальти от агрессии'
f.update_xaxes(title_text = 'Результаты игроков')
f.update_yaxes(title_text = 'Количественный показатель')
f.update_traces()
f.show()
```

Как показывает данная гистограмма - количество назначенных пенальти не всегда зависит от уровня агрессии игрока.
